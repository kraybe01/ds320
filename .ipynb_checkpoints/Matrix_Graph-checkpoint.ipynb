```python
import csv 
%matplotlib inline
import networkx as nx
import warnings
import matplotlib.pyplot as plot
warnings.filterwarnings('ignore')

with open('/home/faculty/millbr02/pub/un-debates-augmented-1.csv','r',encoding='latin-1') as data:
    csv_iteration = csv.reader(data)
    d_base = []
    for row in csv_iteration:
        d_base.append(row)
        
    country_names = {}
    for line in d_base[1:]:
        country_names[line[4]] = 0
   
    matrix = {}
    for row in d_base[1:]:
        speaker = row[4]
        text = row[3]
        if speaker not in matrix:
            matrix[speaker] = country_names.copy()
            
            
        for i in matrix:
            counter = text.count(i)
            matrix[speaker][i] += counter

        
                
                

matrix
```

```python
import matplotlib.pyplot as plt

graf = nx.Graph()

graf.add_nodes_from(matrix) 
for i in matrix: 
    for i2 in matrix[i]: 
        if matrix[i][i2] > 50:
            graf.add_edge(i, i2) 

# position = nx.spring_layout(graf)
# nx.draw(graf, with_labels=True, font_size=10)
gfig,gax = plt.subplots(figsize=(20,20))
pos = nx.spring_layout(graf,k=.3,scale=8.0,center=(0,0))
nx.draw(graf,pos,ax=gax,node_size=1000,with_labels=True,font_size=10)

plt.show()
```
