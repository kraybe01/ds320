```python
import requests
import csv
```

```python
with open('/home/faculty/millbr02/pub/un-debates-augmented-1.csv','r',encoding='latin-1') as data:
    csv_file = csv.reader(data)
    data_base = []
    for row_line in csv_file:
        data_base.append(row_line)

    country_names = {}
    for row_line in data_base[1:]:
        country_names[row_line[4]] = 0
        
    country_ids = {}
    for row_line in data_base[1:]:
        country_ids[row_line[4]] = row_line[2]
        
    temp_list = []
    for row_line in data_base[1:]:
        temp_list.append(row_line[4])
    
    matrix = {}
    for row_line in data_base[1:]:
        speaker = row_line[4]
        speach = row_line[3]
        if speaker not in matrix:
            matrix[speaker] = country_names.copy()
            
        for key_id in matrix:
            counter = speach.count(key_id)
            matrix[speaker][key_id] += counter      
            
    count_mentions = {}
    for key_id in matrix:
        for key_id2 in matrix[key_id]:
            if key_id not in count_mentions:
                count_mentions[key_id] = [int(matrix[key_id][key_id2])]
            else:
                count_mentions[key_id].append(int(matrix[key_id][key_id2]))
                
    count_all = {}        
    for key_id in count_mentions:
        count = sum(count_mentions[key_id])
        count_all[key_id] = count
```
